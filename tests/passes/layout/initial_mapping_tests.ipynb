```python
from QiskitTranspiler.transpiler.passes.layout.floyd_w import FloydWarshall
from qiskit.transpiler import CouplingMap
from qiskit.providers.fake_provider import GenericBackendV2
from QiskitTranspiler.transpiler.passes.layout.layout import Layout
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(3)
qc.cx(0,1)
qc.cx(1,2)
qc.cx(2,0)
```

```python
# 2D grid topology (3x3)
# Qubits arranged as:
#  0 - 1 - 2
#  |   |   |
#  3 - 4 - 5
#  |   |   |
#  6 - 7 - 8

complex_coupling = [
    # Horizontal edges
    [0, 1], [1, 2],
    [3, 4], [4, 5],
    [6, 7], [7, 8],
    # Vertical edges
    [0, 3], [3, 6],
    [1, 4], [4, 7],
    [2, 5], [5, 8],
]

backend_complex = GenericBackendV2(
    num_qubits=9,
    basis_gates=["x", "sx", "rz", "cx"],
    coupling_map=complex_coupling,
    seed=123
)
```

```python
dag = Layout.circuit_to_DAG(qc)
```

```python
print("Coupling Map:", backend_complex.coupling_map)
```

```python
mapping = Layout.get_initial_mapping(dag, backend_complex, qc)
print("Initial mapping:", mapping)
```
