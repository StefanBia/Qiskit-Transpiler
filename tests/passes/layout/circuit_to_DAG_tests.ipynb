```python
from qiskit import QuantumCircuit
from qiskit.providers.fake_provider.generic_backend_v2 import GenericBackendV2
from QiskitTranspiler.transpiler.passes.layout.layout import Layout
```

```python
qc = QuantumCircuit(3)
qc.cx(0,1)
qc.cx(1,2)
qc.cx(2,0)

dag = Layout.circuit_to_DAG(qc)
```

```python
# Print DAG structure
print("=" * 50)
print("DAG STRUCTURE")
print("=" * 50)

# Print nodes
print("\nNodes:")
nodes = dag.get_nodes()
for node in nodes:
    operation = dag.nodes[node]
    print(f"  {node}: {operation}")

# Print edges
print("\nEdges:")
edges = dag.get_edges()
if edges:
    for source, target in edges:
        print(f"  {source} -> {target}")
else:
    print("  No edges")

# Print in-degree and out-degree
print("\nNode dependencies:")
for node in nodes:
    predecessors = dag.get_predecessors(node)
    successors = dag.get_successors(node)
    print(f"  {node}:")
    print(f"    Predecessors: {predecessors if predecessors else 'None'}")
    print(f"    Successors: {successors if successors else 'None'}")
```
