```python
from QiskitTranspiler.transpiler.passes.layout.floyd_w import FloydWarshall
from qiskit.transpiler import CouplingMap
from qiskit.providers.fake_provider import GenericBackendV2

# 2D grid topology (3x3)
# Qubits arranged as:
#  0 - 1 - 2
#  |   |   |
#  3 - 4 - 5
#  |   |   |
#  6 - 7 - 8

complex_coupling = [
    # Horizontal edges
    [0, 1], [1, 2],
    [3, 4], [4, 5],
    [6, 7], [7, 8],
    # Vertical edges
    [0, 3], [3, 6],
    [1, 4], [4, 7],
    [2, 5], [5, 8],
]

backend_complex = GenericBackendV2(
    num_qubits=9,
    basis_gates=["x", "sx", "rz", "cx"],
    coupling_map=complex_coupling,
    seed=123
)

fw_complex = FloydWarshall(backend_complex)

print("Distance matrix for 3x3 grid:")
for i, row in enumerate(fw_complex.dist):
    print(f"Qubit {i}: {row}")


test_paths = [(0, 8), (1, 7), (2, 6), (0, 5)]
print("\nShortest paths:")
for src, dst in test_paths:
    path = fw_complex.get_path(src, dst)
    distance = fw_complex.dist[src][dst]
    print(f"  {src} → {dst}: {path} (distance: {distance})")
```
